#### Obtain the distinct values across `Type 1` and `Type 2`.

Exctract all the values in `Type 1` and `Type 2`. Then create an array containing the distinct values across both fields.

```python
# Extract all values from the "Type 1" and "Type 2" columns
type1_values = pokemon['Type 1'].unique()
type2_values = pokemon['Type 2'].unique()

# Combine unique values from both columns into a single set using the "|" operator
distinct_values = set(type1_values) | set(type2_values)

# Convert the set into a list 
distinct_values_list = list(distinct_values)

# Print the list of distinct values
print(distinct_values_list)
```

#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
pokemon['Name'] = pokemon['Name'].str.replace(r'([A-Za-z]+)Mega', 'Mega', regex=True)

# test transformed data
pokemon.head(10)
```

#### Create a new column called `A/D Ratio` whose value equals to `Attack` divided by `Defense`.

For instance, if a pokemon has the Attack score 49 and Defense score 49, the corresponding `A/D Ratio` is 49/49=1.

```python
pokemon['A/D Ratio'] = pokemon['Attack'] / pokemon['Defense']

# Round the values in the "A/D Ratio" column to two decimal places
pokemon['A/D Ratio'] = pokemon['A/D Ratio'].round(2)
pokemon.head()
```

#### Identify the pokemon with the highest `A/D Ratio`.

```python
highest_ad_ratio_pokemon = pokemon[pokemon['A/D Ratio'] == pokemon['A/D Ratio'].max()]
highest_ad_ratio_pokemon["Name"]
```

#### Identify the pokemon with the lowest A/D Ratio.

```python
lowest_ad_ratio_pokemon = pokemon[pokemon['A/D Ratio'] == pokemon['A/D Ratio'].min()]
lowest_ad_ratio_pokemon["Name"]
```

#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
# Create a new column "Combo Type"
pokemon['Combo Type'] = pokemon['Type 1']  # Initialize with Type 1 values

# Update "Combo Type" based on Type 2 values
pokemon.loc[~pokemon['Type 2'].isna(), 'Combo Type'] = pokemon['Type 1'] + '-' + pokemon['Type 2']

# Display the Pokemon DataFrame with the "Combo Type" column
print(pokemon[['Name', 'Combo Type']])
```

#### Identify the pokemon whose `A/D Ratio` are among the top 5.

```python
top_5_ad_ratio = pokemon.nlargest(5, 'A/D Ratio')
top_5_ad_ratio
```

#### For the 5 pokemon printed above, aggregate `Combo Type` and use a list to store the unique values.

Your end product is a list containing the distinct `Combo Type` values of the 5 pokemon with the highest `A/D Ratio`.

```python
distinct_combo_types = top_5_ad_ratio['Combo Type'].unique()
distinct_combo_types
```

#### For each of the `Combo Type` values obtained from the previous question, calculate the mean scores of all numeric fields across all pokemon.

```python
# Step 1: Initialize a dictionary to store mean scores for each Combo Type
combo_type_means_all_pokemon = {}

# Step 2: Calculate mean scores for each numeric field for each Combo Type
for combo_type in distinct_combo_types:
    combo_type_data = pokemon[pokemon['Combo Type'] == combo_type]
    mean_scores = combo_type_data.mean(numeric_only=True)
    combo_type_means_all_pokemon[combo_type] = mean_scores

# Step 3: Display the unique Combo Types and their mean scores for all Pokémon
for combo_type, mean_scores in combo_type_means_all_pokemon.items():
    print(f"Combo Type: {combo_type}")
    print(mean_scores)
    print("\n")
```
